```python
import matplotlib.pyplot as plt

# Data
years = list(range(1980, 2025))  # Years from 1980 to 2024
inflation = [100.8, 104.5, 164.8, 343.8, 627.5, 672.2, 90.1, 131.3, 343, 3079.5,  # 1980-1989
             2314, 171.7, 24.9, 10.6, 4.2, 3.4, 0.2, 0.5, 0.9, -1.2,             # 1990-1999
             -0.9, -1.1, 25.9, 13.4, 4.4, 9.6, 10.9, 8.8, 8.6, 14.8,             # 2000-2009
             10.5, 9.8, 10.8, 10.6, 23.9, 26.6, 34.3, 25.7, 34.3, 53.5,          # 2010-2019
             42, 48.4, 72.4, 211.4, 193]                                         # 2020-2024

# --- Line Graph (19802024) ---
plt.figure(figsize=(14, 7))
plt.plot(years, inflation, marker='o', linestyle='-', color='blue', label="Inflation Rate")

# Add titles and labels
plt.title(" Argentina Inflation Rate (19802024)", fontsize=16)
plt.xlabel("Years", fontsize=14)
plt.ylabel("Inflation Points", fontsize=14)

# Add grid for better visualization
plt.grid(visible=True, linestyle='--', alpha=0.7)

# Annotate specific points
for i, value in enumerate(inflation):
    if i % 5 == 0 or i == len(inflation) - 1:  # Label every 5th year and the last year
        plt.annotate(f"{value:.1f}", (years[i], inflation[i]),
                     textcoords="offset points", xytext=(0, 10), ha='center')

# Add legend
plt.legend()

# Save and Show the line graph
plt.savefig("detailed_inflation_1980_2024.png")
plt.show()

# --- Detailed Plot (20192024) ---
plt.figure(figsize=(10, 6))
years_focus = list(range(2019, 2024))
inflation_focus = [10.6, 23.9, 26.6, 34.3, 25.7]
plt.plot(years_focus, inflation_focus, marker='o', linestyle='-', color='brown', label="Inflation Rate")

# Add titles and labels
plt.title(" Argentina Inflation Rate (20192024)", fontsize=14)
plt.xlabel("Years", fontsize=12)
plt.ylabel("Inflation Points", fontsize=12)

# Add grid for better visualization
plt.grid(visible=True, linestyle='--', alpha=0.7)

# Set minor ticks and gridlines
plt.minorticks_on()
plt.tick_params(axis='x', which='minor', length=4, width=1, color='gray', labelsize=10)
plt.tick_params(axis='x', which='major', length=6, width=1, color='black', labelsize=10)

# Annotate specific points for emphasis
for i, value in enumerate(inflation_focus):
    plt.annotate(f"{value}", (years_focus[i], inflation_focus[i]),
                 textcoords="offset points", xytext=(0, 5), ha='center')

# Add legend
plt.legend()

# Save and Show the detailed plot
plt.savefig("detailed_inflation_plot.png")
plt.show()

```
