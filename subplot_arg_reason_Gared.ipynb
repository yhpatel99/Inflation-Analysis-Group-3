```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data (replace with actual data)
years = list(range(2013, 2023))

# Data for the plots
gdp = [500, 550, 600, 620, 640, 580, 570, 590, 610, 630]  # in billion USD
crude_oil_imports = [100, 120, 110, 115, 105, 100, 95, 90, 85, 80]  # in thousand barrels per day
refined_petroleum_imports = [2.5, 2.7, 2.6, 2.8, 3.0, 3.2, 3.1, 3.3, 3.5, 3.6]  # in billion USD
m2_money_supply = [5, 6, 7, 8, 9, 10, 12, 15, 18, 22]  # in trillion ARS
exchange_rate = [5, 6, 7, 8, 10, 15, 20, 40, 60, 80]  # ARS per USD
wage_price_index = [50, 52, 55, 60, 68, 80, 95, 110, 125, 140]  # Index
exports = [65, 70, 68, 72, 75, 78, 80, 85, 90, 95]  # in billion USD

# Create a DataFrame
data = pd.DataFrame({
    'Year': years,
    'GDP (Billion USD)': gdp,
    'Crude Oil Imports (kb/d)': crude_oil_imports,
    'Refined Petroleum Imports (Billion USD)': refined_petroleum_imports,
    'M2 Money Supply (Trillion ARS)': m2_money_supply,
    'ARS/USD Exchange Rate': exchange_rate,
    'Wage-Price Index': wage_price_index,
    'Exports (Billion USD)': exports
})

# Set the plot style
sns.set_theme(style="whitegrid")

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(9, 7))

# Plot GDP
sns.lineplot(data=data, x='Year', y='GDP (Billion USD)', ax=axs[0, 0], marker='o', color='blue')
axs[0, 0].set_title('Argentina GDP (2013–2022)')
axs[0, 0].grid(True)

# Plot Crude Oil and Refined Petroleum Imports
sns.lineplot(data=data, x='Year', y='Crude Oil Imports (kb/d)', ax=axs[0, 1], marker='o', label='Crude Oil', color='red')
sns.lineplot(data=data, x='Year', y='Refined Petroleum Imports (Billion USD)', ax=axs[0, 1], marker='x', label='Refined Petroleum', color='green')
axs[0, 1].set_title('Crude Oil and Refined Petroleum Imports (2013–2022)')
axs[0, 1].legend()
axs[0, 1].grid(True)

# Plot M2 Money Supply and Exchange Rate
sns.lineplot(data=data, x='Year', y='M2 Money Supply (Trillion ARS)', ax=axs[1, 0], marker='o', label='M2 Money Supply', color='purple')
sns.lineplot(data=data, x='Year', y='ARS/USD Exchange Rate', ax=axs[1, 0], marker='x', label='ARS/USD Exchange Rate', color='orange')
axs[1, 0].set_title('M2 Money Supply and ARS/USD Exchange Rate (2013–2022)')
axs[1, 0].legend()
axs[1, 0].grid(True)

# Plot Wage-Price Index and Exports
sns.lineplot(data=data, x='Year', y='Wage-Price Index', ax=axs[1, 1], marker='o', label='Wage-Price Index', color='brown')
sns.lineplot(data=data, x='Year', y='Exports (Billion USD)', ax=axs[1, 1], marker='x', label='Exports', color='cyan')
axs[1, 1].set_title('Wage-Price Index and Exports (2013–2022)')
axs[1, 1].legend()
axs[1, 1].grid(True)

# Adjust layout and add a common footer for source citation
fig.tight_layout()
plt.figtext(0.5, -0.05, 'Sources: World Bank, EIA, OEC, CEIC Data, XE.com', wrap=True, horizontalalignment='center', fontsize=10)

# Show the plot
plt.show()
```
